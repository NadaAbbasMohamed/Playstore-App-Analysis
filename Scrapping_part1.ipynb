```python
from matplotlib import pyplot as plt
from selenium import webdriver
from bs4 import BeautifulSoup
import seaborn as sns
import pandas as pd
import numpy as np
import requests
import sqlite3
import urllib
import lxml
import re
```

# Scrap for the apps in the home page:

```python
url = 'https://play.google.com/store/apps'
search_response=requests.get(url)
soup=BeautifulSoup(search_response.content,'html.parser')
```

```python
appBox = soup.findAll("div",{"class":"k6AFYd"})
nameBox = soup.findAll("div",{"class":"bQVA0c"})
appName = soup.findAll("div",{"class":"b8cIId ReQCgd Q9MA7b"})
appCompany = soup.findAll("div",{"class":"b8cIId ReQCgd KoLSrc"})
apprating = soup.findAll("div",{"role":"img"})

len(appBox)
```

```python
app_box = []
name_box = []
app_name = []
app_company = []
app_rating = []

for i in range(len(appBox)):
    app_box.append(appBox[i].text)
    name_box.append(nameBox[i].text)
    app_name.append(appName[i].text)
    app_company.append(appCompany[i].text)
    app_rating.append(apprating[i].get('aria-label').split(" ")[1])
    
rows = []
for row in zip(app_name,app_company,app_rating):
    rows.append(row)
    
df1=pd.DataFrame(data=rows,columns=['app_name', 'app_company', 'app_rating'])
df1.head()
```

```python
appBox = soup.findAll("div",{"class":"Ktdaqe"})
titles = []
i = 0
names = []
for box in appBox:
    title = box.findAll("h2", {"class":"sv0AUd bs3Xnd"})
    appName = box.findAll("div",{"class":"b8cIId ReQCgd Q9MA7b"})
    for app in appName:
        names.append(app.text)
        titles.append(title[0].text)
    i+=1
```

```python
df1['underTitle'] = titles
df1.head()
```

```python
priceBox = soup.findAll("div",{"class":"Z2nl8b"})
priceB = []
price = []
for pBox1 in priceBox:
    priceB.append(pBox1.findAll("span",{"class":"VfPpfd ZdBevf i5DZme"}))
for p in priceB:
    if len(p) != 0 :
        price.append(p.get("span").text)
    else:
        price.append(0)
```

```python
df1['app_price'] = price
df1.head()
```

```python
linkToApp = soup.findAll("a",{"class":"poRVub"})
links = []
for i in linkToApp:
    links.append('https://play.google.com'+i.get('href'))
```

```python
links
```

```python
# get no of comments and gener of each application
gener = []
noOfComments = []

for link in links:
    search_response = requests.get(link)
    soup = BeautifulSoup(search_response.content,'html.parser')
    
    gener.append(soup.findAll("a",{"class":"hrTbp R8zArc", "itemprop":"genre"}))
    noOfComments.append(soup.findAll("span", {"class":"EymY4b"}))
    print("done")
```

```python
geners = []
noOfCom = []
for i in range(len(gener)):
    for g in gener[i]:
        geners.append(g.text)
for c in noOfComments:
    noOfCom.append(int(''.join(c[0].findAll("span")[1].get("aria-label").split(" ")[0].split(','))))
```

```python
for i in range(len(gener)):
    gener[i] = gener[i][0]
df1['gener']=gener

df1['noOfComments'] = noOfCom

page_index = np.ones(len(df1))
df1['page_index'] = page_index
df1['page_index'] = df1['page_index'].astype(int)
df1.head()
```

```python
df1.tail()
```

```python
AppAdditionalInfoList = []
for link in links:
    print('done')
    search_response = requests.get(link)
    soup = BeautifulSoup(search_response.content,'html.parser')
    titles = []
    values = []
    pairs = {}
    title = soup.findAll("div", {"class":"BgcNfc"})
    value = soup.findAll("span",{"class":"htlgb"})
    for t in title:
        titles.append(t.text)

    for v in value:
        values.append(v.text)
    values = values[0::2]

    for t,v in zip(titles, values):
        pairs[t] = v
    AppAdditionalInfoList.append(pairs)
    print('appended')
AppAdditionalInfoList    
```

```python
latestUpdate = []
size = []
Installs =[]
androidVersionSupported = []
InterActiveElem = []
contentRating = []
Report = []
InappPurchases = []

keys = ['Updated','Size','Installs','Requires Android','Content Rating','Report','Interactive Elements','In-app Products']
lists = [latestUpdate, size, Installs, androidVersionSupported, contentRating, Report, InterActiveElem, InappPurchases]
for infolist in AppAdditionalInfoList:
    for k,l in zip(keys,lists):
        l.append(infolist.get(k))
```

```python
df1['latest_Update'] = latestUpdate
df1['size'] = size
df1['Installs'] = Installs
df1['android_Version_Supported'] = androidVersionSupported
df1['InterActive_Elem'] = InterActiveElem
df1['content_Rating'] = contentRating
df1['Report'] = Report
df1['In-app_Purchases'] = InappPurchases

df1.head()
```
