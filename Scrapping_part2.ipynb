```python
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import requests
```

# scrap for the apps in the see more of each section:

```python
url = 'https://play.google.com/store/apps'
search_response=requests.get(url)
soup=BeautifulSoup(search_response.content,'html.parser')
```

```python
SeeMoreButtons = soup.findAll("div",{"class":"W9yFB"})
SeeMoreLinks = []
for button in SeeMoreButtons:
    SeeMoreLinks.append('https://play.google.com'+button.a.get("href"))
```

```python
app_name_total = []
app_company_total = []
app_rating_total = []
under_title_total = []
    
for link in SeeMoreLinks:
    search_response=requests.get(link)
    soup=BeautifulSoup(search_response.content,'html.parser')
    Title = soup.findAll("div",{"class":"xwY9Zc"})[0].h2.text
    
    appBox = soup.findAll("div",{"class":"k6AFYd"})
    nameBox = soup.findAll("div",{"class":"bQVA0c"})
    appName = soup.findAll("div",{"class":"b8cIId ReQCgd Q9MA7b"})
    appCompany = soup.findAll("div",{"class":"b8cIId ReQCgd KoLSrc"})
    apprating = soup.findAll("div",{"role":"img"})
    print(len(apprating))
    print(len(appBox))
    
    app_box = []
    name_box = []
    app_name = []
    app_company = []
    app_rating = []
    under_title = []

    for i in range(len(appBox)):
        app_box.append(appBox[i].text)
        name_box.append(nameBox[i].text)
        app_name.append(appName[i].text)
        app_company.append(appCompany[i].text)
        app_rating.append(apprating[i].get('aria-label').split(" ")[1])
        under_title.append(Title)

    app_name_total.append(app_name)
    app_company_total.append(app_company)
    app_rating_total.append(app_rating)
    under_title_total.append(under_title)
    print("done link")
        
```

```python
len(SeeMoreLinks)
```

```python
app_name = []
app_company = []
app_rating = []
under_title = []

rows = []
for i in range(len(app_name_total)):
    for name, company, rating, title in zip(app_name_total[i], app_company_total[i], app_rating_total[i], under_title_total[i]):
        app_name.append(name)
        app_company.append(company)
        app_rating.append(rating)
        under_title.append(title)

for row in zip(app_name, app_company, app_rating, under_title):
    rows.append(row)
    
df2=pd.DataFrame(data=rows,columns=['app_name', 'app_company', 'app_rating', 'under_title'])
df2.head()
```

```python
app_price = []
for link in SeeMoreLinks:
    search_response = requests.get(link)
    soup = BeautifulSoup(search_response.content,'html.parser')
    priceBox = soup.findAll("div",{"class":"Z2nl8b"})
    priceB = []
    price = []
    for pBox1 in priceBox:
        priceB.append(pBox1.findAll("span",{"class":"VfPpfd ZdBevf i5DZme"}))
    for p in priceB:
        if len(p) != 0 :
            price.append(p[0].span.text)
        else:
            price.append(0)
    app_price.append(price)
```

```python
len(app_price[0])
len(df2)
```

```python
PRICE = []
p1 = []
for pp in app_price:
  p1.append(pp)       
        
for p in p1:
  print(len(p))
  for i in p:
    PRICE.append(i)
df2['app_price'] = PRICE
```

```python
gener_ = []
noOfComments_ = []
j=0

for link_ in SeeMoreLinks:
    search_response = requests.get(link_)
    soup = BeautifulSoup(search_response.content,'html.parser')
    
    linkToApp = soup.findAll("a",{"class":"poRVub"})
    links = []
    for i in linkToApp:
        links.append('https://play.google.com'+i.get('href'))
    
    gener = []
    noOfComments = []

    for link in links:
        search_response = requests.get(link)
        soup = BeautifulSoup(search_response.content,'html.parser')

        gener.append(soup.findAll("a",{"class":"hrTbp R8zArc", "itemprop":"genre"}))
        noOfComments.append(soup.findAll("span", {"class":"EymY4b"}))
        print("inner links done")
    print("---- outer link done ----- ---------------------------------"+str(j))
    j=j+1
    gener_.append(gener)
    noOfComments_.append(noOfComments)
    
```

```python
df2.to_csv('secondPage.csv')
g = pd.Series(gener_)
n = pd.Series(noOfComments_)

g.to_csv('gener.csv')
n.to_csv('noOfComments.csv')
```

```python
all_links = []
for link_ in SeeMoreLinks:
    search_response = requests.get(link_)
    soup = BeautifulSoup(search_response.content,'html.parser')
    
    linkToApp = soup.findAll("a",{"class":"poRVub"})
    for i in linkToApp:
        all_links.append('https://play.google.com'+i.get('href'))

all_links
```

```python
l = pd.Series(all_links)
l.to_csv('links.csv')

v = pd.read_csv('links.csv', index_col=0)
all_links = v.values
```

```python
all_links
```

```python
AppAdditionalInfoList = []
i=0
for link in all_links:
    print('done')
    search_response = requests.get(link[0])
    soup = BeautifulSoup(search_response.content,'html.parser')
    titles = []
    values = []
    pairs = {}
    title = soup.findAll("div", {"class":"BgcNfc"})
    value = soup.findAll("span",{"class":"htlgb"})
    for t in title:
        titles.append(t.text)

    for v in value:
        values.append(v.text)
    values = values[0::2]

    for t,v in zip(titles, values):
        pairs[t] = v
    AppAdditionalInfoList.append(pairs)
    print('appended'+ str(i))
    i+=1
```

```python
latestUpdate = []
size = []
Installs =[]
androidVersionSupported = []
InterActiveElem = []
contentRating = []
Report = []
InappPurchases = []

keys = ['Updated','Size','Installs','Requires Android','Content Rating','Report','Interactive Elements','In-app Products']
lists = [latestUpdate, size, Installs, androidVersionSupported, contentRating, Report, InterActiveElem, InappPurchases]
for infolist in AppAdditionalInfoList:
    for k,l in zip(keys,lists):
        l.append(infolist.get(k))
```
